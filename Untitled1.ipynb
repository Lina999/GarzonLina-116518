```python
#PUNTO 6

num_=[] #lista vacía para almacenar digitos
a=0 #variables iniciales fundamentales para fibonacci
b=1#variables iniciales fundamentales para fibonacci
while True:
    try:
        num=int(input("Digite un numero:"))# ingresar numero 
        break
        
    except:# asegurar que sea un numero
        print("Asegúrese de que escrbió un número")
      
if num<0:
    num=num*-1 # si es negativo obtener valor absoluto, Fibonacci es positiva

        
while num!=0:   #mientras tenga digitos
    x=num%10    # separar cada digito
    num_.append(x)   #agrgar a una lista vacía
    num=num//10           #  pasar al siguiente digito
  

y=sum(num_) #suma de los digitos del numero ingresado
print('la suma es:',y)

while y>=0: #Si la suma es mayor o igual a cero
    a,b=b,a+b # a será b y b será el resultado de sumar a y b
    if a>=y: #hacer hasta que a sea maypor o igual a y y parar
        break 
        

if a==y: # si los resultados de las dos sumas mencionadas son iguales, el \
#numero digitado pertenece a la serie de Fibonacci
    print('Pertenece a fibonacci')
# Imprimir resuktado de si pertenece o  no a Fibonacci    
else:
    print('El número no pertenece a fibonacci')
        
   
    
```
